```python
# importing libraries
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler, StandardScaler

from sklearn.cluster import KMeans
import joblib
import json
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read dataset in dataframe
df = pd.read_csv("Driving Data(KIA SOUL)_(150728-160714)_(10 Drivers_A-J).csv")
```

```python
df = df[['Fuel_consumption','Absolute_throttle_position', 'Engine_speed', 'Current_Gear', 'Steering_wheel_angle',
         'Steering_wheel_speed','Acceleration_speed_-_Longitudinal','Calculated_LOAD_value', 'Engine_coolant_temperature',
        'Activation_of_Air_compressor','Vehicle_speed', 'Indication_of_brake_switch_ON/OFF','Class']]
```

```python
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'Country'. 
df['Class']= label_encoder.fit_transform(df['Class']) 
```

```python
scaler = StandardScaler()

# Fit and transform the data
fit_data = scaler.fit(df)
normalized_data = fit_data.transform(df)
# Create a new DataFrame with the normalized data
df = pd.DataFrame(normalized_data, columns=df.columns)
```

```python
joblib.dump(fit_data, 'fit_data.save')
```
